### Define and Plot a Function

To carry out the graphing of a function, we will follow these steps:

1. Define a function
2. Create an array of $x$-values to be plugged into $f(x)$
3. Plot the array of $x$-values against the values $f(x)$

For our example, let us plot and graph the functions from class 1:

1. $l(x) = 40x + 6$
2. $s(x) = 10\sqrt{(x-3)^2 + 7}$
3. $t(x) = 30 + 30 \sin{18x}$
4. $e(x) = \frac{2^x}{1000}$

````python
def l(x):
    return 40*x + 6

def s(x):
    return 10*np.sqrt((x-3)**2+7)

def t(x):
    return 3 + 3*np.sin(1.8*x)

def e(x):
    return (2**(2*x))/(10)
````

````python
x = np.linspace(0,100,1000)
````

````python
plt.figure(figsize=(12,8))
plt.plot(x,l(x), label = '$l(x)$')
plt.plot(x,s(x), label = '$s(x)$')
plt.plot(x, t(x), label = '$t(x)$')
plt.plot(x,e(x), label = '$e(x)$')
plt.xlim(0,5)
plt.ylim(0,100)
plt.legend(loc='best')
plt.title('All Four on One Graph')
````

#### Subplots

In the example below, we will plot each function on their own as a 2 by 2 grid.  The logic is as follows:

1. First we create a figure with: `fig = plt.rigure()`
2. Then we add subplots to this figure one by one with `fig.add_subplot(221)` command means that we have 2 rows, 2 columns, and this is the first sublplot graph.  

The following are rearranged accordingly.  You can then add customizations etc. to the subplots.

````python
fig = plt.figure(figsize=(12,10))

ax1 = fig.add_subplot(221)
ax1.plot(x,l(x), 'r-')

ax2 = fig.add_subplot(222)
ax2.plot(x, s(x), 'k-')

ax3 = fig.add_subplot(223)
ax3.plot(x, t(x), 'b-')

ax4 = fig.add_subplot(224)
ax4.plot(x, e(x), 'g-')
````

## Define and Plot Sequences

While there are many ways to approach the construction of sequences, we can follow a similar path to what we have above.  Remember, the difference here is that we won't know the **closed form** of the sequence, and likely define it recursively.  Two examples follow.

1. Suppose we start with \$1 and add \$1/day for 30 days.
2. Suppose we start with \$0.01 and double it every day for 30 days.

````python
a = 1
for i in range(1,31):
    a = a + 1
    print(i, a)
````

````python
b = .01
for i in range(1,31):
    b = b *2
    print(i, b)
````

**PLOTTING**: To plot these we can simply add another line to each of these codes to add the plot.  Below is an example of each sequence plotted on the same axes.  Play around with the number of terms, limits, scales, and customizations to get a nice final graph.

````python
b = 0.01
a = 1
for i in range(1,12):
    b = b*2
    a = a + 1
    plt.plot(i,b, marker = 'o')
    plt.plot(i,a,marker = 'o')
````
