### Prompt I


##### What is consumer surplus?
Consumer surplus depicts the amount of money consumers "save" by pursuing a commodity at a certain price P (decided by producers), instead of purchasing the commodity at the maximum price the consumer is willing to pay for. In graphs, consumer surplus is the area between the demand curve and the line y = P, P usually representing the equilibrium price.


##### What does consumer surplus have to do with integration?
Integration is connected to consumer surplus as consumer surplus is identified as the area between the demand curve and the line y = P, while integration is commonly used to calculate areas below and between curves.


##### What are some assumptions that we make about supply and demand curves when investigating these ideas?
When investigating these ideas, we are making several assumptions about the supply and demand curves. Firstly, we assume that the curves are generalizable -- when, in fact, the interactions within a market are so diverse that one simply model cannot possible be applicable to all situations. Secondly, we assume of a competitive market in which there are infinite amount of consumers and producers. In this case, the action of one consumer or one producer cannot affect the curve. Lastly, we assume that all consumers and all producers have perfect knowledge of the quantity/availability of the commodity in the market.

### Prompt II


##### What is a Lorenz curve?
Lorenz curve is a curve which demonstrates the equality/inequality of income distribution of a country with an interval [0, 1]. A point [x, y] on the curve states that the poorest 100x% of the population receives 100y% of the total income.


##### What is the Gini Index?
The Gini Index is the area between line y = x and the Lorenz curve divided by the area under y = x (which is 0.5) .


##### What are some assumptions we make about distribution functions when investigating these ideas?
When investigating thesse ideas, we make several assumptions about distribution functions. Firstly, we assume that all wealth of a citizen is solely based on income. This is not realistically accurate as people may have material goods as wealth or have an underground source of wealth. However, these forms of wealth would be difficult to quantify. Secondly, we assume that population is finite. Lastly, we assume that error/uncertainty of data is negligible/nonexistent.

Lorenz curves describe the distribution of income within a percentage of a country's population. For example, point (0.6, 0.2) on the green Lorenz curve illustrates that the poorer 60% of that country's population receives only 20% of the total income. The steepest line y = x demonstrates the Lorenz curve of a perfectly egalitarian society. Hence, the steeper the Lorenz curve, the more equal a country's total income is distributed. A less steep Lorenz curve indicates more inequality within a country's income distribution.

```python
fig, ax = plt.subplots()

x = np.linspace(0, 1, 100)
f = x
g = x**3
plt.plot(x, f, color = 'black', lw=2.5)
plt.fill_between(x, f, g, where = (x>=0), color='#F0FFF0')
plt.plot(x, g, color = 'blue', lw = 2.5)

plt.grid(True)
plt.axis('equal')
plt.ylabel('income fraction')
plt.xlabel('population fraction')

plt.plot([1, 1], [0, 1], color='k', linestyle='--', linewidth=1)

ax.annotate('y = x', (0.45, 0.6), size=10)
ax.annotate('y = L(x)', (0.63, 0.2), size=10, color='blue')

plt.show

plt.title('Explain the Gini Index and draw a diagram to explain this.', y=1.08, fontsize=15)
```

Gini index is used as a manner to quantify a country's inequality in income distribution. A smaller number
indicates less inequality in income distribution and vice versa. In a graph, the Gini Index is the area between line y = x and the Lorenz curve y = L(x) -- the shaded area -- divided by the area under y = x (which is 0.5) .


##### Explain and derive the equation $$G = 2 \int_0^1 [x − L(x)] dx$$

Gini index is the area between line y = x and the Lorenz curve divided by the area under y = x. The area between line y = x and Lorenz curve L(x) -- from x = 0 to x = 1 as population cumulates to 100% -- can be calculated using integral 

$$\int_0^1 [x − L(x)] dx$$

The area under y = x is (1/2) or 0.5 as y = x cuts the 1x1 square into 2. Hence,

 $$G = \frac{\int_0^1 [x − L(x)] dx}{1/2} = 2 \int_0^1 [x − L(x)] dx$$
